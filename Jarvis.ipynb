```python
import speech_recognition as sr
import pyttsx3
import datetime
import wikipedia
import webbrowser
import os
#import wolframalpha
import requests
```

```python
# Initialize text-to-speech engine
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # Change index for different voices

def speak(audio):
    """Convert text to speech"""
    engine.say(audio)
    engine.runAndWait()

def take_command():
    """Takes microphone input and returns text"""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source)
        
    try:
        print("Recognizing...")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")
        
    except Exception as e:
        print("Say that again please...")
        return "None"
    return query.lower()

def greet():
    """Greets the user based on time"""
    hour = datetime.datetime.now().hour
    if 0 <= hour < 12:
        speak("Good Morning Sir")
    elif 12 <= hour < 18:
        speak("Good Afternoon Sir")
    else:
        speak("Good Evening Sir")
    speak("Jarvis at your service. How may I assist you?")

# def get_weather(city="new york"):
#     """Get weather information"""
#     api_key = "YOUR_OPENWEATHER_API_KEY"
#     base_url = "http://api.openweathermap.org/data/2.5/weather?"
#     complete_url = f"{base_url}appid={api_key}&q={city}"
    
#     response = requests.get(complete_url)
#     data = response.json()
    
#     if data["cod"] != "404":
#         main = data["main"]
#         temperature = main["temp"] - 273.15
#         weather = data["weather"][0]["description"]
#         speak(f"Temperature is {temperature:.1f} degrees Celsius with {weather}")
#     else:
#         speak("City not found")
```
